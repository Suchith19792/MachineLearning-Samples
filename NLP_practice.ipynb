```python
import spacy_annotator as spa
import pandas as pd
from pprint import pprint
```

```python
text_split = sample_text.split('\n')
text_split = [item for item in text_split]
print(len(text_split))
print(text_split)
```

```python
df = pd.DataFrame(text_split,columns=['text'])
df.head()
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
annotator = spa.Annotator(labels=['company','person','money','space','date'],model=nlp)
```

```python
df_labels = annotator.annotate(df=df,col_text='text')
```

```python
df_labels['annotations'][0]
```

```python
df_labels.to_csv('annotated_labels.csv',index=False)
```
